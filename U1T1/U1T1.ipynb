**Exercise 2.2:**

60% of the emails hitting my inbox is spam. You design a phenomenal spam filter which is able to tell me, with 98% accuracy, whether an email is spam or not: if an email is not spam, the system has a 98% probability of saying so. The filter knows 60% of emails are spam and so it will flag 60% of my emails. Suppose that, at the end of the week, I look in my spam box and see 963 emails. Use Bayes’ Theorem to calculate how many of those 963 emails in my spam box I should suspect to be non-spam.

**Solution:**

The algorithm starts by defining the known probabilities: P(Spam) = 0.6 and P(-Spam) = 0.4. We know that P(-Flagged|-Spam) = 0.98, so P(Flagged|-Spam) = 0.02. With P(Flagged) = 0.6, we apply Bayes' Theorem to calculate P(-Spam|Flagged): P(-Spam|Flagged) = (P(Flagged|-Spam) × P(-Spam)) / P(Flagged) = (0.02 × 0.4) / 0.6 = 0.01333. We multiply this value by the total number of emails in the spam folder (963) to obtain 963 × 0.01333 = 12.84, indicating that about 13 emails are not spam.

**Exercise 2.3:**

You're given the string: "OCZ XJMMZXO VINRZM". Each letter follows a stochastic Markov process with the rules expressed by this table. Follow the process for three steps and reconstruct the correct answer. (Note, this is a Caesar cipher with shift 7 applied three times, because the Caesar cipher is a Markov process).

**Solution:**

The code implements a basic Caesar cipher decryption function that shifts letters in a given text by a specified amount. It defines separate alphabets for uppercase and lowercase letters and processes each character, applying the shift to obtain the decrypted result. The initial text is decrypted in three stages, with the output of each stage printed. This approach demonstrates how repeated decryption can be used to retrieve original text from multiple layers of encryption.


```python
# Exercise 2.3
def basic_caesar_decrypt(text, shift):
    uppercase_alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    lowercase_alphabet = 'abcdefghijklmnopqrstuvwxyz'
    result = ''
    
    for char in text:
        if char in uppercase_alphabet:
            index = uppercase_alphabet.index(char)
            new_index = (index - shift) % 26
            result += uppercase_alphabet[new_index]
        elif char in lowercase_alphabet:
            index = lowercase_alphabet.index(char)
            new_index = (index - shift) % 26
            result += lowercase_alphabet[new_index]
        else:
            result += char  
    return result


initial_text = "OCZ XJMMZXO VINRZM"
shift = 7


first_stage_text = basic_caesar_decrypt(initial_text, shift)
print("1st Stage:", first_stage_text)


second_stage_text = basic_caesar_decrypt(first_stage_text, shift)
print("2nd Stage:", second_stage_text)


third_stage_text = basic_caesar_decrypt(second_stage_text, shift)
print("3rd Stage:", third_stage_text)

```

**Exercise 2.4:**

Draw the probability mass function and the cumulative distribution of the following outcome probabilities:

| Outcome | p |
|---|---|
| 1 | 0.1 |
| 2 | 0.15 |
| 3 | 0.2 |
| 4 | 0.21 |
| 5 | 0.17 |
| 6 | 0.09 |
| 7 | 0.06 |
| 8 | 0.02 |

**Solution:**

The code creates and visualizes probability distributions using a DataFrame. It defines outcomes and their corresponding probabilities, then calculates the cumulative probabilities. Two bar charts are generated: the first displays the Probability Mass Function (PMF), showing the individual probabilities for each outcome, and the second illustrates the Cumulative Distribution Function (CDF), depicting the cumulative probabilities. This visualization effectively highlights the distribution of outcomes and their likelihoods.

```python
#Exercise 2.4
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Outcome': [1, 2, 3, 4, 5, 6, 7, 8],
    'Probability': [0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02]
}

df = pd.DataFrame(data)


df['Cumulative Probability'] = df['Probability'].cumsum()

plt.figure(figsize=(10, 5))

plt.bar(df['Outcome'], df['Probability'], color='red', alpha=0.7, label='p', width=0.4, align='center')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass Function')
plt.xticks(df['Outcome']) 
plt.legend()
plt.grid(axis='y')
plt.show()

plt.figure(figsize=(10, 5))

plt.bar(df['Outcome'], df['Cumulative Probability'], color='red', alpha=0.7, label='cumulative', width=0.4, align='center')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function')
plt.xticks(df['Outcome']) 
plt.legend()
plt.grid(axis='y')
plt.show()
```

**Exercise 3.2:**

Mr. A is friend with Mrs. B, but she doesn't like him back. She has a reciprocal friendship with both C and D, but only C considers D a friend. D has also sent friend requests to E, F, G, and H but, so far, only G replied. G also has a reciprocal relationship with A. Draw the corresponding directed graph.

**Solution:**

The code creates a directed graph using NetworkX and adds edges to define its structure. It calculates the in-degree and out-degree for each vertex, displaying the results. The shortest path from vertex A to vertex C is identified. The code also finds and prints the strongly connected components of the graph. Additionally, it computes the degree centrality and betweenness centrality for all vertices, providing insights into their importance within the network. Finally, the graph is visualized using Matplotlib, displaying the connections between vertices.

```python
# Exercise 3.2:
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()
G.add_edges_from([('A', 'B'), ('B', 'D'), ('C', 'D'), ('D', 'A'), ('D', 'B'),
                  ('D', 'C'), ('D', 'E'), ('D', 'F'), ('D', 'G'), ('D', 'H')])

in_degree_dict = dict(G.in_degree())
out_degree_dict = dict(G.out_degree())
print("In-degree:")
print(in_degree_dict)
print("\nOut-degree:")
print(out_degree_dict)

shortest_path = nx.shortest_path(G, source='A', target='C')
print("Shortest path from A to C:", shortest_path)

strongly_connected_components = nx.strongly_connected_components(G)
print("Strongly connected components:")
for component in strongly_connected_components:
    print(component)

degree_centrality = nx.degree_centrality(G)
print("Degree centrality:")
print(degree_centrality)

betweenness_centrality = nx.betweenness_centrality(G)
print("Betweenness centrality:")
print(betweenness_centrality)  

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.show()
```

**Exercise 3.3:**

Draw the graph from Exercise 3.2 (image below) as undirected and weighted, with the weight being 2 if the connection is reciprocal, 1 otherwise.

**Solution:**

The code creates an undirected graph using NetworkX, adding nodes labeled from A to H. It establishes edges between the nodes, each with a specified weight. The graph is visualized using a spring layout, displaying the nodes and their connections. Edge labels indicate the weights, providing insights into the relationships between the nodes. Finally, the graph is displayed using Matplotlib, illustrating the network's structure and weights clearly.

```python
#Exercise 3.3 

import networkx as nx
import matplotlib.pyplot as plt


G = nx.Graph()


G.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])


G.add_edge('A', 'B', weight=1)
G.add_edge('A', 'G', weight=2)
G.add_edge('B', 'C', weight=2)
G.add_edge('B', 'D', weight=2)
G.add_edge('C', 'D', weight=2)
G.add_edge('D', 'E', weight=1)
G.add_edge('D', 'F', weight=1)
G.add_edge('D', 'G', weight=2)
G.add_edge('D', 'H', weight=1)
G.add_edge('E', 'H', weight=1)
G.add_edge('F', 'G', weight=1)

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)  

plt.show()
```

**Exercise 3.4:**

Draw a correlation network for these vectors, by only drawing edges with positive weights, ignoring self loops.

**Solution:**

The code constructs a correlation network from a DataFrame of vectors, focusing on positive correlations. It calculates the correlation matrix and filters for positive weights, ignoring self-loops. The edges representing these correlations are then visualized using NetworkX and Matplotlib, effectively highlighting the relationships between the variables in the dataset. The resulting graph displays only the connections with positive correlation values.

```python
#Exercise 3.4


import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

data = {
    "A": [1, 1, 1, 1, 1, 2, 3, 4, 4, 3, 5, 6, 7, 8, 9],
    "B": [6, 4, 4, 4, 5, 2, 6, 6, 4, 6, 4, 6, 4, 6, 7],
    "C": [1, 4, 4, 6, 7, 5, 7, 9, 8, 7, 7, 9, 7, 0, 3],
    "D": [1, 6, 6, 8, 7, 7, 0, 5, 4, 3, 2, 4, 5, 2, 9],
    "E": [1, 1, 1, 1, 2, 2, 3, 3, 4, 5, 5, 6, 7, 7, 8],
    "F": [3, 5, 5, 7, 0, 9, 7, 3, 2, 0, 8, 8, 4, 2, 3],
    "G": [1, 3, 7, 8, 0, 9, 6, 3, 3, 3, 7, 7, 5, 4, 5],
    "H": [1, 7, 8, 4, 5, 5, 4, 7, 5, 5, 5, 4, 7, 3, 6]
}
df = pd.DataFrame(data)
df = df.corr()

df = df > 0
df = df.unstack().reset_index()
df.columns = ("source", "target", "edge")
df = df[df["edge"]]
df = df[df["source"] != df["target"]]
G = nx.from_pandas_edgelist(df)
nx.draw(G, with_labels = True)
plt.show()
```
